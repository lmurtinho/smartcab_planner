# Training a Smartcab Planner

This notebook presents the results of my implementation of a learning planner for the smartcab world presented in project 4 of Udacity's Machine Learning Nanodegree.

## Module Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import random
from IPython.display import display
```

## Random Seed

```python
random.seed(1)
```

## Matplotlib Parameters

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10]
plt.rcParams['axes.titlesize'] = 24
plt.rcParams['axes.labelsize'] = 22
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['legend.fontsize'] = 18
```

## Object Imports

```python
from planner_world import PlannerWorld
from planner_agents import RandomPlanner, PerfectPlanner, LearningPlanner
from run_functions import sim_single_planner, sim_all_planners
```

```python
GRID = (8, 6)
TRIALS = 2000
DEADLINE = 100
PLANNERS = [RandomPlanner(), PerfectPlanner(), LearningPlanner(),
            LearningPlanner(learn_rate_func = lambda x: 1.0 / np.log(x + 1))]
PLANNERS_NAMES = ['random', 'perfect', 'original learner', 'final learner']

results, qvals = sim_all_planners(GRID, TRIALS, DEADLINE, PLANNERS, PLANNERS_NAMES)
```

```python
for planner in results:
    df_planner = results[planner]
    print "- Results for {} planner:".format(planner)
    print "- Destinations reached: {} of {}".format(sum(df_planner['n_steps']<DEADLINE), TRIALS)
    print "- Positive sums of rewards: {} of {}".format(sum(df_planner['reward_sum']>0), TRIALS)
    print "- States visited: {}".format(len(qvals[planner]))
    display(df_planner.describe())
    print "\n\n\n"
```

```python
panel = pd.Panel(results)

for column in ['reward_sum', 'n_steps']:
    panel.minor_xs(column).rolling(window=100, center=False).mean().plot()
    plt.title("Rolling mean of {} over {} trials".format(column, TRIALS))
    plt.xlabel("Trial")
    plt.ylabel(column)
    plt.savefig("roll_mean_{}.png".format(column))
    plt.show()
    plt.close()
```
